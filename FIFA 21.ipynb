## Extracting some basic columns below

```python
# Create new pandas DataFrame
df_fifa = df[['ID','Name','Age','Gender','Club','Height','Weight','Joined','Value','Wage','Release Clause','Hits','Team & Contract']]
df_fifa.head()
```

## Check for missing values

```python
df_fifa.isnull().sum()
```

```python
df_fifa[['Team & Contract', 'Club', 'Joined' ]].head(30)
```

#### Observations 
1. Team & Contract column contains some information which is similar to the values in Club and Joined columns.
2. There are players that are currently not associated with any specific club, therefore we can drop these players as they are more suited to be labelled under different category.

```python
# Drop rows in which players are labelled as free players
df_fifa = df_fifa[~df_fifa['Team & Contract'].str.contains('Free', na=False)]

# Count missing values in the data frame
df_fifa.isnull().sum()
```

```python
import numpy as np

# Extract the first set of years from 'Team & Contract' column (only if it's exactly four digits)
extracted_years = df_fifa['Team & Contract'].str.findall(r'\b(\d{4})\b').str[0]

# Impute missing values with NaT if no set of four digits (year) is found
df_fifa['Joined'] = np.where(extracted_years.isnull(), pd.NaT, extracted_years)
```

```python
df_fifa.isnull().sum()
```

```python
# Show rows with NaT values in 'Joined' column
null_rows = df_fifa[df_fifa['Joined'].isnull()][['Joined', 'Team & Contract']]
null_rows
```

```python
df_fifa = df_fifa.dropna(subset=['Joined'])
```

## Check data types of Height and Weight

```python
print(df_fifa[['Height','Weight']].dtypes)

# Top 5 rows of Height and Weight
df_fifa[['Height','Weight',]].head()
```

```python
# Change Height into numerical variable

def convert_to_feet(height_str):
  feet, inches = height_str.split("'")
  feet = int(feet) # Convert to numeric
  inches = float(inches.strip('\"'))
  total_height_feet = round((feet + inches / 12), 2)

  return total_height_feet

# Transform
df_fifa['Height'] = df_fifa['Height'].apply(convert_to_feet)
```

```python
import re
# Function to convert weight string to numeric value
def convert_weight(weight_str):
    numeric_weight = re.sub(r'\D', '', weight_str)
    return int(numeric_weight)

# Apply the conversion function to the 'Weight' column
df_fifa['Weight'] = df_fifa['Weight'].apply(convert_weight)

# Rename column 'Weight' to 'Weight(lbs)'
df_fifa = df_fifa.rename(columns={'Weight': 'Weight(lbs)'})

# Show the data types
df_fifa[['Height','Weight(lbs)']].dtypes
```

## Check data types of Value, Wage and Release Clause

```python
# Show the data types
df_fifa[['Value','Wage','Release Clause']].dtypes

```

```python
# Top 5 rows of 'Value', 'Wage' and 'Release Clause'
df_fifa[['Value','Wage','Release Clause']].head(5)
```

```python
# Function to convert string value to integer
def convert_money(money_str):
    multiplier = 1  # Default multiplier for values without 'K' or 'M' suffix

    if money_str.endswith('K'):
        multiplier = 1000
        money_str = money_str[:-1]  # Remove the 'K' suffix
    elif money_str.endswith('M'):
        multiplier = 1000000
        money_str = money_str[:-1]  # Remove the 'M' suffix

    money_str = money_str[1:]  # Remove the euro symbol

    return int(float(money_str) * multiplier) # Return integer value
```

```python
# Apply the conversion function to the 'Value' , 'Wage' and 'Release Clause' column
df_fifa['Value'] = df_fifa['Value'].apply(convert_money)
df_fifa['Wage'] = df_fifa['Wage'].apply(convert_money)
df_fifa['Release Clause'] = df_fifa['Release Clause'].apply(convert_money)
```

```python
# Show the data types
df_fifa[['Value','Wage','Release Clause']].dtypes
```

## Convert the 'Joined' column to datetime type

```python
df_fifa['Joined'] = pd.to_datetime(df_fifa['Joined'])

# Extract year, month, and day components into separate columns
df_fifa['Year'] = df_fifa['Joined'].dt.year.astype(int)
df_fifa['Month'] = df_fifa['Joined'].dt.month.astype(int)
df_fifa['Day'] = df_fifa['Joined'].dt.day.astype(int)

# Top 5 rows of 'Joined', 'Year', 'Month' and 'Day' columns
df_fifa[['Joined','Year','Month','Day']].head()
```
